```python
size = (dataset.shape[0]//3)
print(size)
```

```python
#Assigning labels according to the happiness index into three classes
# 0:Happy
# 1:Neutral
# 2:sad
dataset['Class'] = 0
dataset.iloc[size:2*size]['Class'] = 1
dataset.iloc[2*size:]['Class'] = 2
```

```python
dataset.head(20)
```

```python
# Happiest countries among each group
dataset.iloc[[0, size, 2*size], :]
```

```python
def distplot(col, bins=10):
    
    fig, ax = plt.subplots(1,3,figsize=(16, 4))

    sns.distplot(dataset[dataset['Class']==0][col], bins=10, ax=ax[0])
    ax[0].set_title('Class 0')

    sns.distplot(dataset[dataset['Class']==1][col], bins=10, ax=ax[1])
    ax[1].set_title('Class 1')

    sns.distplot(dataset[dataset['Class']==2][col], bins=10, ax=ax[2])
    ax[2].set_title('Class 2')

    plt.show();
```

```python
def scatterplot(x, y):
    fig, ax = plt.subplots(1, 3, figsize=(16, 6))
    sns.regplot(x, y, data=dataset[dataset['Class']==0], ax=ax[0])
    ax[0].set_title('Class 0', size=15)
    ax[0].set_xlabel(x, size=15)
    ax[0].set_ylabel(y, size=15)
    sns.regplot(x, y, data=dataset[dataset['Class']==1], ax=ax[1])
    ax[1].set_title('Class 1')
    ax[1].set_title('Class 1', size=15)
    ax[1].set_xlabel(x, size=15)
    ax[1].set_ylabel(y, size=15)
    sns.regplot(x, y, data=dataset[dataset['Class']==2], ax=ax[2])
    ax[2].set_title('Class 2')
    ax[2].set_title('Class 2', size=15)
    ax[2].set_xlabel(x, size=15)
    ax[2].set_ylabel(y, size=15)
    plt.show()
```
