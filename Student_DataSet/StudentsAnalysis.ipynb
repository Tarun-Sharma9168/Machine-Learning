```python
dataset['status'] = dataset.apply(lambda x : 'Fail' if x['pass_math'] == 'Fail' or 
                           x['pass_reading'] == 'Fail' or x['pass_writing'] == 'Fail'
                           else 'pass', axis = 1)
dataset['status'].value_counts(dropna = False).plot.pie(colors = ['grey', 'crimson'])
plt.title('overall results', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
```

```python
def get_grade(percentage,status):
    if (status == 'Fail'):
        return 'E'
    if (percentage >= 90) :
        return 'O'
    if (percentage >= 80) :
        return 'A'
    if (percentage >= 70) :
        return 'B'
    if (percentage >= 60) :
        return 'C'
    if (percentage >= 40) :
        return 'D'
    else :
        return 'E'
dataset['grades'] = dataset.apply(lambda x: get_grade(x['percentage'], x['status']),axis=1)    
dataset['grades'].value_counts()
```

```python
#start building the machine learning model 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["test preparation course"]=le.fit_transform(dataset["test preparation course"])
```

```python
dataset["test preparation course"]
```

```python
dataset["lunch"].head()
```

```python
#Now change the encoding of the lunch attribute also
dataset["lunch"]=le.fit_transform(dataset["lunch"])
dataset["lunch"].head()
```

```python
dataset["status"]
```

```python
# label encoding for race/ethnicity
# we have to map values to each of the categories
dataset['race/ethnicity'] = dataset['race/ethnicity'].replace('group A', 1)
dataset['race/ethnicity'] = dataset['race/ethnicity'].replace('group B', 2)
dataset['race/ethnicity'] = dataset['race/ethnicity'].replace('group C', 3)
dataset['race/ethnicity'] = dataset['race/ethnicity'].replace('group D', 4)
dataset['race/ethnicity'] = dataset['race/ethnicity'].replace('group E', 5)
```

```python
# label encoding for parental level of education
dataset['parental level of education'] = le.fit_transform(dataset['parental level of education'])

#label encoding for gender
dataset['gender'] = le.fit_transform(dataset['gender'])

# label encoding for pass_math
dataset['pass_math'] = le.fit_transform(dataset['pass_math'])

# label encoding for pass_reading
dataset['pass_reading'] = le.fit_transform(dataset['pass_reading'])

# label encoding for pass_writing
dataset['pass_writing'] = le.fit_transform(dataset['pass_writing'])

# label encoding for status
dataset['status'] = le.fit_transform(dataset['status'])
```

```python
#preparation of the data
X = dataset.iloc[:,:14].values
y = dataset.iloc[:,14].values

print(X.shape)
print(y.shape)
```

```python
#splitting the model into the training and the test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#You can print the shape of the data that you make 
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
```

```python
#Implemennting Logistics Regression Classifier
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
```

```python
#printing the confusion matrices and check if some false positives and false negatives examples are there or not
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
# printing the confusion matrix
plt.rcParams['figure.figsize'] = (8, 8)
ax=sns.heatmap(cm, annot = True, cmap = 'Greens')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.title('Confusion Matrix for Logistic Regression', fontweight = 30, fontsize = 20)
plt.show()
```

```python
#printing the accuracy of the model
print(classifier.score(X_train,y_train)*100,"%")
print(classifier.score(X_test,y_test)*100,"%")
```

```python
#Now trying the random forest clasifier
from sklearn.ensemble import RandomForestClassifier
classfier=RandomForestClassifier()
classfier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
```
